```python
df = pd.read_csv(r"C:\Users\Guicheney William\Documents\GitHub\Challenges\Challenge-003-Running-Averages\files\input.csv")
df['RecordID'] = df.index
```

```python
df.head(5)
```

```python
df = df.melt(id_vars=['RecordID', 'HP Category', 'Year', 'Month'],
            var_name='Measure',
            value_name='Value')
df
```

```python
df['total'] = 0
df['MovingAverage']=0

for index, n in enumerate(df):
    if index == 0:
        df['total'] = df['Value']
        df['MovingAverage'] = df['total']/1
    elif index == 1:
        df['total'] = df['Value'] + df['Value'][index-1]
        df['MovingAverage'] = df['total']/2
    else:
        df['total'] = df['Value'] + df['Value'][index-1] + df['Value'][index-2]
        df['MovingAverage'] = df['total']/3
    
```
