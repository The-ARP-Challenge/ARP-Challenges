# Running Averages

The goal is to create a 3 and 6 month running averages for the values contained in columns: 
u.CAGI, d.CAGI, u.IR, d.IR, u.NonIR, d.NonIR by HP Category.

__*For example:*__


|Month:|Value:|Running 3 mo. Average:|
|---|---|---|
|1|218|218|
|2|200|212|
|3|272|233|

*Hint:* For Values that do not exist, set values to closest valid row.

```python
# Import required Packages

import pandas as pd
```

```python
# read in input dataset
df_input = pd.read_csv('../files/input.csv')
df_input.head()
```

```python
#read in output dataset for comparasion
df_output = pd.read_csv('../files/output.csv', index_col='RecordID')
print(df_output.shape)
df_output.head()
```

```python
df_input.shape
```

```python
# Columns to create running totals for
#cols = ['u.CAGI', 'd.CAGI', 'u.IR', 'd.IR', 'u.NonIR', 'd.NonIR']
cols = ['d.CAGI', 'd.IR', 'd.NonIR', 'u.CAGI', 'u.IR', 'u.NonIR']

# initialise empty output dataframe
df = pd.DataFrame
for col in cols:
    label = "".join(['r3mo_', col.replace('.', '_')])
    df_input[label] = df_input.groupby('HP Category')[col].rolling(3,min_periods=1).mean().reset_index('HP Category', drop =True)
    
df = pd.DataFrame
for col in cols:
    label = "".join(['r6mo_', col.replace('.', '_')])
    df_input[label] = df_input.groupby('HP Category')[col].rolling(6,min_periods=1).mean().reset_index('HP Category', drop =True)

print(df_input.shape)
df_input.reset_index
df_input.head()
```

```python
#(df_input == df_output).all()

col_match = (df_input.columns == df_output.columns)
col_no_match = (df_input.columns != df_output.columns)
print(col_match)
print(df_input.columns)
print(df_input.columns[col_no_match])
print(df_output.columns[col_no_match])
```
